```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Данные (те же, что и в части 1)
X_data = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
y_and = torch.tensor([[0.], [0.], [0.], [1.]])
y_or  = torch.tensor([[0.], [1.], [1.], [1.]])
y_xor = torch.tensor([[0.], [1.], [1.], [0.]])

print("X_data shape:", X_data.shape)
print("y_and shape:", y_and.shape)
```

```python
# Модель через Sequential
model_and_seq = nn.Sequential(
    nn.Linear(2, 1),
    nn.Sigmoid()
)

# Функция потерь и оптимизатор
criterion = nn.MSELoss()
optimizer = optim.SGD(model_and_seq.parameters(), lr=0.5, momentum=0.9)

# Обучение
epochs = 500
loss_history = []

for epoch in range(epochs):
    # Прямой проход
    outputs = model_and_seq(X_data)
    loss = criterion(outputs, y_and)
    loss_history.append(loss.item())
    
    # Обратный проход и оптимизация
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 100 == 0:
        print(f'Epoch {epoch+1}, Loss: {loss.item():.6f}')

# Проверка точности
with torch.no_grad():
    pred = (model_and_seq(X_data) >= 0.5).float()
    acc = (pred == y_and).float().mean().item()
    print(f'Accuracy: {acc:.2f}')
```

```python
plt.plot(loss_history)
plt.xlabel('Эпоха')
plt.ylabel('Потеря (MSE)')
plt.title('Снижение потерь при обучении AND')
plt.grid(True)
plt.show()
```

```python
class OneLayerNetwork(nn.Module):
    def __init__(self, input_size, output_size):
        super(OneLayerNetwork, self).__init__()
        self.linear = nn.Linear(input_size, output_size)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        return self.sigmoid(self.linear(x))

# Использование для OR
model_or_class = OneLayerNetwork(2, 1)
criterion = nn.MSELoss()
optimizer = optim.SGD(model_or_class.parameters(), lr=0.5, momentum=0.9)

# Обучение (аналогично)
epochs = 500
for epoch in range(epochs):
    outputs = model_or_class(X_data)
    loss = criterion(outputs, y_or)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Проверка
with torch.no_grad():
    pred = (model_or_class(X_data) >= 0.5).float()
    acc = (pred == y_or).float().mean().item()
    print(f'OR accuracy (class model): {acc:.2f}')
```

```python
class TwoLayerNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(TwoLayerNetwork, self).__init__()
        self.hidden = nn.Linear(input_size, hidden_size)
        self.output = nn.Linear(hidden_size, output_size)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        hidden_out = self.sigmoid(self.hidden(x))
        out = self.sigmoid(self.output(hidden_out))
        return out

# Инициализация (скрытый слой из 2 нейронов)
model_xor = TwoLayerNetwork(2, 2, 1)
criterion = nn.MSELoss()
optimizer = optim.SGD(model_xor.parameters(), lr=0.7, momentum=0.9)

# Обучение
epochs = 1000
loss_history_xor = []

for epoch in range(epochs):
    outputs = model_xor(X_data)
    loss = criterion(outputs, y_xor)
    loss_history_xor.append(loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 200 == 0:
        print(f'Epoch {epoch+1}, Loss: {loss.item():.6f}')

# Проверка точности
with torch.no_grad():
    pred = (model_xor(X_data) >= 0.5).float()
    acc = (pred == y_xor).float().mean().item()
    print(f'XOR accuracy: {acc:.2f}')
```

```python
plt.plot(loss_history_xor)
plt.xlabel('Эпоха')
plt.ylabel('Потеря (MSE)')
plt.title('Снижение потерь при обучении XOR')
plt.grid(True)
plt.show()
```

```python
def train_and_evaluate(model_class, optimizer_name, lr=0.5, momentum=0.9, epochs=500):
    # Создаём модель и оптимизатор
    model = model_class(2, 1)
    criterion = nn.MSELoss()
    
    if optimizer_name == 'SGD':
        optimizer = optim.SGD(model.parameters(), lr=lr, momentum=momentum)
    elif optimizer_name == 'Adam':
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name == 'RMSprop':
        optimizer = optim.RMSprop(model.parameters(), lr=lr)
    else:
        raise ValueError("Unknown optimizer")
    
    loss_history = []
    epoch_acc = []
    
    for epoch in range(epochs):
        outputs = model(X_data)
        loss = criterion(outputs, y_or)
        loss_history.append(loss.item())
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # Проверка точности каждые 10 эпох
        if (epoch+1) % 10 == 0:
            with torch.no_grad():
                pred = (model(X_data) >= 0.5).float()
                acc = (pred == y_or).float().mean().item()
                epoch_acc.append((epoch+1, acc))
                if acc == 1.0:
                    print(f"{optimizer_name}: достигнута 100% точность на эпохе {epoch+1}")
                    break
    return loss_history, epoch_acc

# Сравнение
opt_names = ['SGD', 'Adam', 'RMSprop']
histories = {}
for opt in opt_names:
    print(f"\n--- {opt} ---")
    loss_hist, acc_hist = train_and_evaluate(OneLayerNetwork, opt, lr=0.5, epochs=500)
    histories[opt] = loss_hist
    # График потерь
    plt.plot(loss_hist, label=opt)
    
plt.xlabel('Эпоха')
plt.ylabel('Потеря (MSE)')
plt.legend()
plt.title('Сравнение оптимизаторов для задачи OR')
plt.show()
```
