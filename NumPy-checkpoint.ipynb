```python
X_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and_data = np.array([0, 0, 0, 1]).reshape(-1, 1)

np.random.seed(21)
w = np.random.uniform(low=-0.05, high=0.05, size=(2, 1))
b = np.random.uniform(low=-0.05, high=0.05)
```

```python
# MSE
def mse(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)

def mse_derivative(y_pred, y_true):
    return 2 * (y_pred - y_true) / len(y_true)   # усреднённая производная

# Бинарная кросс-энтропия
def bce(y_pred, y_true):
    eps = 1e-12
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def bce_derivative(y_pred, y_true):
    eps = 1e-12
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return (y_pred - y_true) / (y_pred * (1 - y_pred) + eps) / len(y_true)

def sigmoid(s, c=1):
    return 1. / (1. + np.exp(-c*s))

def sigmoid_derivative(s):
    z = sigmoid(s)
    return z * (1 - z)

def mse(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)

def mse_derivative(y_pred, y_true):
    return 2 * (y_pred - y_true) / len(y_true)
```

```python
def train_neuron(X, y_true, loss_func, loss_deriv, activate, activate_deriv,
                 w, b, epochs, lr, momentum):
    velocity_w = np.zeros_like(w)
    velocity_b = 0.0
    loss_history = []
    for epoch in range(epochs):
        # Прямой проход
        s = X.dot(w) + b
        z = activate(s)
        # Потеря
        loss = loss_func(z, y_true)
        loss_history.append(loss)
        # Градиенты
        dL_dz = loss_deriv(z, y_true)
        dz_ds = activate_deriv(s)
        dL_ds = dL_dz * dz_ds
        dL_dw = X.T.dot(dL_ds)
        dL_db = np.sum(dL_ds)
        # Обновление с импульсом
        velocity_w = momentum * velocity_w + lr * dL_dw
        velocity_b = momentum * velocity_b + lr * dL_db
        w -= velocity_w
        b -= velocity_b
    return w, b, loss_history

# Параметры
epochs = 200
learning_rate = 0.5
momentum = 0.9

# Инициализация
np.random.seed(21)
w = np.random.uniform(-0.05, 0.05, (2,1))
b = np.random.uniform(-0.05, 0.05)

# Обучение с MSE
w_and, b_and, loss_history_mse = train_neuron(
    X_data, y_and_data, mse, mse_derivative, sigmoid, sigmoid_derivative,
    w.copy(), b, epochs, learning_rate, momentum)

# Оценка точности
s = X_data.dot(w_and) + b_and
z = sigmoid(s)
pred = (z >= 0.5).astype(int)
acc = np.mean(pred == y_and_data)
print(f"MSE: accuracy = {acc:.2f}, final loss = {loss_history_mse[-1]:.6f}")
```

```python
# Обучение с BCE
np.random.seed(21)
w = np.random.uniform(-0.05, 0.05, (2,1))
b = np.random.uniform(-0.05, 0.05)
w_and_bce, b_and_bce, loss_history_bce = train_neuron(
    X_data, y_and_data, bce, bce_derivative, sigmoid, sigmoid_derivative,
    w, b, epochs, learning_rate, momentum)

s = X_data.dot(w_and_bce) + b_and_bce
z = sigmoid(s)
pred = (z >= 0.5).astype(int)
acc_bce = np.mean(pred == y_and_data)
print(f"BCE: accuracy = {acc_bce:.2f}, final loss = {loss_history_bce[-1]:.6f}")

# Построим графики потерь
plt.plot(loss_history_mse, label='MSE')
plt.plot(loss_history_bce, label='BCE')
plt.xlabel('Эпоха')
plt.ylabel('Потеря')
plt.legend()
plt.title("Сравнение MSE и BCE для AND")
plt.show()
```

```python
y_xor_data = np.array([0, 1, 1, 0]).reshape(-1, 1)

def train_xor_network(n_hidden, epochs, lr, momentum, 
                      act1, act1_deriv, act2, act2_deriv):
    np.random.seed(42)
    w1 = np.random.randn(2, n_hidden) * 0.5
    b1 = np.random.randn(1, n_hidden) * 0.5
    w2 = np.random.randn(n_hidden, 1) * 0.5
    b2 = np.random.randn(1, 1) * 0.5
    
    vel_w1 = np.zeros_like(w1)
    vel_b1 = np.zeros_like(b1)
    vel_w2 = np.zeros_like(w2)
    vel_b2 = np.zeros_like(b2)
    
    loss_history = []
    for epoch in range(epochs):
        # Прямой проход
        s1 = X_data.dot(w1) + b1
        z1 = act1(s1)
        s2 = z1.dot(w2) + b2
        z2 = act2(s2)
        
        loss = mse(z2, y_xor_data)
        loss_history.append(loss)
        
        # Обратный проход
        dL_dz2 = mse_derivative(z2, y_xor_data)
        dz2_ds2 = act2_deriv(s2)
        dL_ds2 = dL_dz2 * dz2_ds2
        
        dL_dw2 = z1.T.dot(dL_ds2)
        dL_db2 = np.sum(dL_ds2, axis=0, keepdims=True)
        
        dL_dz1 = dL_ds2.dot(w2.T)
        dz1_ds1 = act1_deriv(s1)
        dL_ds1 = dL_dz1 * dz1_ds1
        
        dL_dw1 = X_data.T.dot(dL_ds1)
        dL_db1 = np.sum(dL_ds1, axis=0, keepdims=True)
        
        # Обновление с импульсом
        vel_w1 = momentum * vel_w1 + lr * dL_dw1
        vel_b1 = momentum * vel_b1 + lr * dL_db1
        vel_w2 = momentum * vel_w2 + lr * dL_dw2
        vel_b2 = momentum * vel_b2 + lr * dL_db2
        
        w1 -= vel_w1
        b1 -= vel_b1
        w2 -= vel_w2
        b2 -= vel_b2
        
    # Финальная точность
    s1 = X_data.dot(w1) + b1
    z1 = act1(s1)
    s2 = z1.dot(w2) + b2
    z2 = act2(s2)
    pred = (z2 >= 0.5).astype(int)
    acc = np.mean(pred == y_xor_data)
    return acc, loss_history
```

```python
mom_list = [0.0, 0.5, 0.9]
for mom in mom_list:
    np.random.seed(21)
    w = np.random.uniform(-0.05, 0.05, (2,1))
    b = np.random.uniform(-0.05, 0.05)
    _, _, loss_hist = train_neuron(
        X_data, y_and_data, mse, mse_derivative, sigmoid, sigmoid_derivative,
        w, b, 200, 0.5, mom)
    print(f"momentum={mom}: final loss = {loss_hist[-1]:.6f}")
    plt.plot(loss_hist, label=f'mom={mom}')
plt.legend()
plt.title("Влияние momentum на сходимость")
plt.show()
```

```python
for h in [2, 3, 5]:
    acc, hist = train_xor_network(h, epochs=500, lr=0.5, momentum=0.9,
                                   act1=sigmoid, act1_deriv=sigmoid_derivative,
                                   act2=sigmoid, act2_deriv=sigmoid_derivative)
    print(f"Hidden neurons = {h}: accuracy = {acc:.2f}")
    plt.plot(hist, label=f'{h} нейронов')
plt.legend()
plt.title("Влияние числа нейронов скрытого слоя на сходимость XOR")
plt.show()
```
